```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual


    
```

```python
quanQual(dataset)
```

```python
quan,qual=quanQual(dataset)
```

```python
descriptive_table=pd.DataFrame(index=["Mean","Median","Mode"],columns=quan)
for columnName in quan:
        descriptive_table[columnName]["Mean"]=dataset[columnName].mean()
        descriptive_table[columnName]["Median"]=dataset[columnName].median()
        descriptive_table[columnName]["Mode"]=dataset[columnName].mode()[0]
```

```python
descriptive_table=pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%','Q3:75%','Q4:100%','IQR','1.5rule','lesser','greater','min','max'],columns=quan)
for col in descriptive_table.columns:
    descriptive_table[col]['Mean']=dataset[col].mean()
    descriptive_table[col]['Median']=dataset[col].median()
    descriptive_table[col]['Mode']=dataset[col].mode()[0]
    descriptive_table[col]['Q1:25%']=dataset.describe()[col]['25%']
    descriptive_table[col]['Q2:50%']=dataset.describe()[col]['50%']   
    descriptive_table[col]['Q3:75%']=dataset.describe()[col]['75%']
    descriptive_table[col]['Q4:100%']=dataset.describe()[col]['max']
    descriptive_table[col]['IQR']=descriptive_table[col]['Q3:75%']-descriptive_table[col]['Q1:25%']
    descriptive_table[col]['1.5rule']=1.5*descriptive_table[col]['IQR']
    descriptive_table[col]['lesser']= descriptive_table[col]['Q1:25%']-descriptive_table[col]['1.5rule']
    descriptive_table[col]['greater']= descriptive_table[col]['Q3:75%']+descriptive_table[col]['1.5rule']
    descriptive_table[col]['min']=dataset[col].min()
    descriptive_table[col]['max']=dataset[col].max()

    
    
    
```
